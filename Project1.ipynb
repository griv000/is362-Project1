## Airline arrival delays
Analyzing data for arrival delays for two airlines across five destinations.

```python
# import modules
import numpy as np
import pandas as pd

# import matplotlib for visuals
import matplotlib.pyplot as pyplot
%matplotlib inline
```

Import the csv table. The information has been reformatted to transform it into a standard spreadsheet

```python
airline_table = pd.read_csv('Project1.csv')
airline_table
```

This shows the cumulative minutes that each airline's flights where on time and delayed:

```python
airline_table.groupby(['Airline','Status']).sum().sort_values(by=['Status'], ascending=False)
```

This shows the breakdown of on time vs delayed arrivals to each city:

```python
airline_table.groupby(['Status','City']).sum().sort_values(by=['Status','Minutes'],ascending=False)
```

The following is a visualization of the arrival delays to all cities:

```python
# Creating element which only shows flights with status of Delayed
airline_table2 = airline_table.query('Status == "delayed"')

# Creating element which groups those delayed flights by city and sorts in descending order

airline_table3 = airline_table2.groupby(['City']).sum().sort_values(by=['Minutes'], ascending=False)
airline_table3

# Visualizing Data
airline_table3.plot.bar()
pyplot.ylabel("Minutes")
pyplot.show()
```
